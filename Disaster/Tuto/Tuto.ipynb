```python
# import data about hurricane
florence = pd.read_csv('./tuto/florence.csv');
florence.head()
```

```python
florence.info()
```

```python
# number of data
msn.bar(florence, color='darkolivegreen');
```

```python
florence.describe()
```

```python
# drop the row with missing value
florence = florence.drop(['AdvisoryNumber','Forecaster', 'Received'], axis=1)
florence.head()
```

```python
# change into proper
florence['Long'] = 0 - florence['Long']
florence.head()
```

```python
# combine long and lat
florence['coordinates'] = florence[['Long','Lat']].values.tolist()
florence.head()
```

```python
# change type to Point
florence['coordinates'] = florence['coordinates'].apply(Point)
florence.head()
```

```python
type(florence)
```

```python
type(florence['coordinates'])
```

```python
florence = geopandas.GeoDataFrame(florence, geometry='coordinates')
florence.head()
```

```python
type(florence)
```

```python
type(florence['coordinates'])
```

```python
print("Mean wind spped {} mph\nMax wind speed {} mph".format(round(florence.Wind.mean(),4),florence.Wind.max()))
```

```python
florence.plot(figsize=(20,10))
```

```python
# hurricane overlay over US map
fig, ax = plt.subplots(1, figsize=(30,20))
base = country[country['NAME'].isin(['Alaska', 'Hawaii']) == False ].plot(ax=ax, color='#3B3C6E')

# color markers
florence.plot(ax=base, color='darkred', marker='*', markersize=20)
```

```python
# add title
# color position based on speed
# remove axis
# add legend
fig,ax = plt.subplots(1, figsize=(50,20))
base = country[country['NAME'].isin(['Alaska', 'Hawaii']) == False].plot(ax=ax, color='#3B3C6E')
florence.plot(ax=base, column='Wind', marker='<', markersize=20, cmap='cool', label = 'Wind speed(mph)', legend=True)
_ = ax.axis('off')
ax.set_title('Hurricane Florence in US Map', fontsize = 25)
plt.savefig('Hurricane_footage.png', bbox_inches='tight')
```
